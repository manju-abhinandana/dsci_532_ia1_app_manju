```python
def plot_altair(region, year):
    chart = alt.Chart(gapminder[gapminder['region'] == region] & gapminder[gapminder['year'] == year]).mark_point().encode(
        x = 'life_expectancy',
        y = 'income',
        color='country',
        size='population').properties(title = "Life expectancy vs income")
    return chart.to_html()

```
